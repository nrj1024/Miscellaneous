# With wave module

```python
import matplotlib.pyplot as plt
import wave
import numpy as np
```

```python
audio = wave.open("hello.wav", "r")       #has to be a mono channel .wav
buffer = audio.readframes(-1)
data = np.frombuffer(buffer, dtype='int16')
```

```python
plt.figure(1)
plt.title("Waveform")
plt.plot(data)
plt.show()
```

# With scipy module

```python
import matplotlib.pyplot as plt
from scipy.io import wavfile as wav
data = wav.read('hello.wav')[1]            #returns [samplerate, array, datatype]
plt.plot(data)
plt.show()
```
